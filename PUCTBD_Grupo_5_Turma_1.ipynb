```python
import pandas as pd
from datetime import date

# Converter as consultas (queries) presentes na especificação de requisitos funcionais:

# Q01 - Identificar os 3 produtos mais vendidos:

# SELECT Nome,
#       SUM(Quantidade) AS `Total de Vendas`
# FROM Item
# JOIN Produto ON Produto_Código = Código
# GROUP BY Produto_Código
# ORDER BY `Total de Vendas` DESC
# LIMIT 3;

df_q1 = df_item.merge(df_produto, left_on='Produto_Codigo', right_on='Codigo', how='inner')
df_q1 = df_q1.groupby('Produto').agg({'Quantidade': 'sum'}).reset_index()
df_q1 = df_q1.sort_values('Quantidade', ascending=False).head(3)


# Q02 - Identificar os 3 produtos menos vendidos

# SELECT Nome,
#       SUM(Quantidade) AS `Total de Vendas`
# FROM Item
# JOIN Produto ON Produto_Código = Código
# GROUP BY Produto_Código
# ORDER BY `Total de Vendas`
# LIMIT 3;

df_q2 = df_item.merge(df_produto, left_on='Produto_Codigo', right_on='Codigo', how='inner')
df_q2 = df_q2.groupby('Produto').agg({'Quantidade': 'sum'}).reset_index()
df_q2 = df_q2.sort_values('Quantidade').head(3)

# Q03 - Visualizar os 3 pedidos mais caros

# SELECT Número AS `Número do Pedido`,
#      CONCAT('R$ ', FORMAT(SUM(Quantidade * Valor), 2)) AS `Valor Total`
# FROM Pedido
# JOIN Item ON Número = Pedido_Número
# JOIN Produto pr ON Produto_Código = Código
# GROUP BY Número
# ORDER BY SUM(Quantidade * Valor) DESC
# LIMIT 3;

df_q3 = df_pedido.merge(df_item, left_on='Numero', right_on='Pedido_Numero', how='inner')
df_q3 = df_q3.merge(df_produto, left_on='Produto_Codigo', right_on='Codigo', how='inner')
df_q3['Valor Total'] = df_q3['Quantidade'] * df_q3['Valor']
df_q3 = df_q3.groupby('Numero').agg({'Valor Total': 'sum'}).reset_index()
df_q3 = df_q3.sort_values('Valor Total', ascending=False).head(3)
df_q3['Valor Total'] = df_q3['Valor Total'].apply(lambda x: f'R$ {x:.2f}')



# Q04 - Visualizar o ticket médio de compras

# SELECT CONCAT('R$ ', ROUND(AVG(`Valor Total`), 2)) AS `Ticket Médio`
# FROM (
#  SELECT Número AS `Número do Pedido`,
#           SUM(Quantidade * Valor) AS `Valor Total`
#  FROM Pedido
#  JOIN Item ON Número = Pedido_Número
#  JOIN Produto ON Produto_Código = Código
#  GROUP BY Número
#) AS subquery;



df_q4 = df_pedido.merge(df_item, left_on='Numero', right_on='Pedido_Numero')
df_q4 = df_q4.merge(df_produto, left_on='Produto_Codigo', right_on='Codigo')
df_q4['Valor Total'] = df_q4['Quantidade'] * df_q4['Valor']
df_q4 = df_q4.groupby('Numero').agg({'Valor Total': 'sum'}).reset_index()
df_q4 = df_q4['Valor Total'].mean()
df_q4 = 'R$ {:.2f}'.format(df_q4)

# Q05 - Identificar os 2 clientes que mais compram

# SELECT Nome,
#       COUNT(`Cliente_ClienteID`) AS Compras
# FROM Pedido
# JOIN Cliente ON ClienteID = Cliente_ClienteID
# GROUP BY Cliente_ClienteID
# ORDER BY Compras DESC
# LIMIT 2;

df_q5 = df_pedido.merge(df_cliente, left_on='ClienteID', right_on='ClienteID')
df_q5 = df_q5.groupby('Nome').size().reset_index(name='Compras')
df_q5 = df_q5.sort_values('Compras', ascending=False).head(2)

# Q06 - Identificar os 2 clientes que menos compram

# SELECT Nome,
#       COUNT(`Cliente_ClienteID`) AS Compras
# FROM Pedido
# JOIN Cliente ON ClienteID = Cliente_ClienteID
# GROUP BY Cliente_ClienteID
# ORDER BY Compras
# LIMIT 2;

df_q6 = df_pedido.merge(df_cliente, left_on='ClienteID', right_on='ClienteID')
df_q6 = df_q6.groupby('Nome').size().reset_index(name='Compras')
df_q6 = df_q6.sort_values('Compras').head(2)

# Q07 - Visualizar idade média de fãs

# SELECT CONCAT(ROUND(AVG(TIMESTAMPDIFF(YEAR, `Data de Nascimento`, CURDATE()))), ' anos') AS IdadeMedia
# FROM Cliente;

df_cliente['Data de Nascimento'] = pd.to_datetime(df_cliente['Data de Nascimento'])
df_cliente['Idade'] = (pd.Timestamp.now().normalize() - df_cliente['Data de Nascimento']).astype('<m8[Y]')
idade_media = round(df_cliente['Idade'].mean())
df_q7 = f"{idade_media} anos"

# Q08 - Visualizar preferências de compra por faixa etária

# SELECT
#   CASE
#    WHEN TIMESTAMPDIFF(YEAR, `Data de Nascimento`, CURDATE()) BETWEEN 18 AND 25 THEN '18-25'
#    WHEN TIMESTAMPDIFF(YEAR, `Data de Nascimento`, CURDATE()) BETWEEN 26 AND 35 THEN '26-35'
#    WHEN TIMESTAMPDIFF(YEAR, `Data de Nascimento`, CURDATE()) BETWEEN 36 AND 45 THEN '36-45'
#    ELSE '46+'
#   END AS FaixaEtaria,
#  Produto.Nome AS Produto,
#  COUNT(*) AS `Total de Pedidos`
# FROM  Cliente
# JOIN  Pedido  ON ClienteID = Cliente_ClienteID
# JOIN  Item ON Número = Pedido_Número
# JOIN  Produto ON Produto_Código = Código
# GROUP BY  FaixaEtaria, Produto
# ORDER BY  FaixaEtaria, `Total de Pedidos` DESC;

faixas_etarias = {
    (18, 25): '18-25',
    (26, 35): '26-35',
    (36, 45): '36-45',
    (46, float('inf')): '46+'
}
df_cliente['Idade'] = pd.to_datetime(df_cliente['Data de Nascimento']).apply(lambda x: date.today().year - x.year)
df_cliente['FaixaEtaria'] = pd.cut(df_cliente['Idade'], bins=[18, 26, 36, 46, float('inf')], right=False, labels=[faixas_etarias[key] for key in faixas_etarias])
df_q8 = df_cliente.merge(df_pedido, left_on='ClienteID', right_on='ClienteID')
df_q8 = df_q8.merge(df_item, left_on='Numero', right_on='Pedido_Numero')
df_q8 = df_q8.merge(df_produto, left_on='Produto_Codigo', right_on='Codigo')
df_q8 = df_q8.groupby(['FaixaEtaria', 'Produto']).size().reset_index(name='Total de Pedidos')
df_q8 = df_q8.sort_values(['FaixaEtaria', 'Total de Pedidos'], ascending=[True, False])

# Q09 - Visualizar preferências de compra por sexo

# SELECT Sexo,
#       Produto.Nome AS Produto,
#        COUNT(*) AS `Total de Pedidos`
# FROM Cliente
# JOIN Pedido ON ClienteID = Cliente_ClienteID
# JOIN Item ON Número = Pedido_Número
# JOIN Produto  ON Produto_Código = Código
# GROUP BY Produto, Sexo
# ORDER BY Sexo, `Total de Pedidos` DESC;

df_q9 = df_cliente.merge(df_pedido, left_on='ClienteID', right_on='ClienteID')
df_q9 = df_q9.merge(df_item, left_on='Numero', right_on='Pedido_Numero')
df_q9 = df_q9.merge(df_produto, left_on='Produto_Codigo', right_on='Codigo')
df_q9 = df_q9.groupby(['Sexo', 'Produto']).size().reset_index(name='Total de Pedidos')
df_q9 = df_q9.sort_values(['Sexo', 'Total de Pedidos'], ascending=[True, False])

# Q10 - Visualizar hobbies mais comuns entre fãs

# SELECT Nome AS Hobby,
#       COUNT(*) AS `Total de Clientes`
# FROM Hobby
# JOIN `Cliente/Hobby` ON HobbyID = Hobby_HobbyID
# GROUP BY Nome
# ORDER BY `Total de Clientes` DESC;

df_q10 = df_hobby.merge(df_cliente_hobby, left_on='HobbyID', right_on='Hobby_HobbyID')
df_q10 = df_q10.groupby('Nome').size().reset_index(name='Total de Clientes')
df_q10 = df_q10.sort_values('Total de Clientes', ascending=False)

# Q11 - Visualizar total de compras por cidade

# SELECT Cidade,
#       COUNT(*) AS `Total de Clientes`
# FROM Cliente
# JOIN Pedido ON ClienteID = Cliente_ClienteID
# GROUP BY Cidade
# ORDER BY `Total de Clientes` DESC;

df_q11 = df_cliente.merge(df_pedido, left_on='ClienteID', right_on='ClienteID')
df_q11 = df_q11.groupby('Cidade').size().reset_index(name='Total de Clientes')
df_q11 = df_q11.sort_values('Total de Clientes', ascending=False)


# Q12 - Visualizar total de vendas por período

# SELECT YEAR(`Data`) AS `Ano`,
#       MONTH(`Data`) AS `Mês`,
#       SUM(`Quantidade`) AS `Total de Vendas`
# FROM Pedido
# JOIN Item ON Número = Pedido_Número
# GROUP BY YEAR(`Data`), MONTH(`Data`)
# ORDER BY YEAR(`Data`), MONTH(`Data`);

df_q12 = df_pedido.merge(df_item, left_on='Numero', right_on='Pedido_Numero')
df_q12['Ano'] = pd.DatetimeIndex(df_q12['Data']).year
df_q12['Mês'] = pd.DatetimeIndex(df_q12['Data']).month
df_q12 = df_q12.groupby(['Ano', 'Mês']).agg({'Quantidade': 'sum'}).reset_index()
df_q12 = df_q12.sort_values(['Ano', 'Mês'])

# Q13 - Visualizar valor de vendas por período

# SELECT YEAR(`Data`) AS `Ano`,
#       MONTH(`Data`) AS `Mês`,
#       CONCAT('R$ ', FORMAT(SUM(`Quantidade` * `Valor`),2)) AS `Total de Vendas`
# FROM Pedido
# JOIN Item ON Número = Pedido_Número
# JOIN Produto ON Produto_Código = Código
# GROUP BY YEAR(`Data`), MONTH(`Data`)
# ORDER BY YEAR(`Data`), MONTH(`Data`);

df_q13 = df_pedido.merge(df_item, left_on='Numero', right_on='Pedido_Numero')
df_q13 = df_q13.merge(df_produto, left_on='Produto_Codigo', right_on='Codigo')
df_q13['Ano'] = pd.DatetimeIndex(df_q13['Data']).year
df_q13['Mês'] = pd.DatetimeIndex(df_q13['Data']).month
df_q13['Total de Vendas'] = df_q13['Quantidade'] * df_q13['Valor']
df_q13 = df_q13.groupby(['Ano', 'Mês']).agg({'Total de Vendas': 'sum'}).reset_index()
df_q13 = df_q13.sort_values(['Ano', 'Mês'])
df_q13['Total de Vendas'] = df_q13['Total de Vendas'].apply(lambda x: f'R$ {x:.2f}')


# Q14 - Visualizar fornecedor com mais vendas

# SELECT CNPJ,
#       Nome,
#        SUM(Quantidade) AS `Total de Vendas`
# FROM Fornecedor
# JOIN `Fornecedor/Produto` ON CNPJ = Fornecedor_CNPJ
# JOIN Item ON `Fornecedor/Produto`.Produto_Código = Item.Produto_Código
# GROUP BY CNPJ, Nome
# ORDER BY SUM(Quantidade) DESC;

df_q14 = df_fornecedor.merge(df_fornecedor_produto, left_on='CNPJ', right_on='Fornecedor_CNPJ')
df_q14 = df_q14.merge(df_item, left_on='Produto_Codigo', right_on='Produto_Codigo')
df_q14 = df_q14.groupby(['CNPJ', 'Nome']).agg({'Quantidade': 'sum'}).reset_index()
df_q14 = df_q14.sort_values('Quantidade', ascending=False)

# Q15 - Visualizar fornecedor mais rentável

# SELECT Fornecedor.Nome AS Fornecedor,
#       SUM(Quantidade * Valor) AS `Receita Total`
# FROM Fornecedor
# JOIN `Fornecedor/Produto` ON CNPJ = Fornecedor_CNPJ
# JOIN Produto ON Produto_Código = Código
# JOIN Item ON Código = item.Produto_Código
# JOIN Pedido ON Pedido_Número = Número
# GROUP BY Fornecedor
# ORDER BY `Receita Total` DESC;

df_q15 = df_fornecedor.merge(df_fornecedor_produto, left_on='CNPJ', right_on='Fornecedor_CNPJ')
df_q15 = df_q15.merge(df_produto, left_on='Produto_Codigo', right_on='Codigo')
df_q15 = df_q15.merge(df_item, left_on='Codigo', right_on='Produto_Codigo')
df_q15 = df_q15.merge(df_pedido, left_on='Pedido_Numero', right_on='Numero')
df_q15['Receita Total'] = df_q15['Quantidade'] * df_q15['Valor']
df_q15 = df_q15.groupby('Nome').agg({'Receita Total': 'sum'}).reset_index()
df_q15 = df_q15.sort_values('Receita Total', ascending=False)


# Q16 - Visualizar total de clientes por fornecedor

# SELECT Nome AS Fornecedor, COUNT(Cliente_ClienteID) AS `Total de Clientes`
# FROM Fornecedor
# JOIN `Cliente/Fornecedor` ON CNPJ = Fornecedor_CNPJ
# GROUP BY Nome
# ORDER BY `Total de Clientes` DESC;

df_q16 = df_fornecedor.merge(df_cliente_fornecedor, left_on='CNPJ', right_on='Fornecedor_CNPJ')
df_q16['Total de Clientes'] = df_q16['Cliente_ClienteID']
df_q16 = df_q16.groupby('Nome').agg({'Total de Clientes': 'count'}).reset_index()
df_q16 = df_q16.sort_values('Total de Clientes', ascending=False)


print('Q01 - Identificar os 3 produtos mais vendidos')
print('')
print(df_q1.to_string(index=False))
print('')
print('Q02 - Identificar os 3 produtos menos vendidos')
print('')
print(df_q2.to_string(index=False))
print('')
print('Q03 - Visualizar os 3 pedidos mais caros')
print('')
print(df_q3.to_string(index=False))
print('')
print('Q04 - Visualizar o ticket médio de compras')
print('')
print('Ticket Médio:', df_q4)
print('')
print('Q05 - Identificar os 2 clientes que mais compram')
print('')
print(df_q5.to_string(index=False))
print('')
print('Q06 - Identificar os 2 clientes que menos compram')
print('')
print(df_q6.to_string(index=False))
print('')
print('Q07 - Visualizar idade média de fãs')
print('')
print(df_q7)
print('')
print('Q08 - Visualizar preferências de compra por faixa etária')
print('')
print(df_q8.to_string(index=False))
print('')
print('Q09 - Visualizar preferências de compra por sexo')
print('')
print(df_q9.to_string(index=False))
print('')
print('Q10 - Visualizar hobbies mais comuns entre fãs')
print('')
print(df_q10.to_string(index=False))
print('')
print('Q11 - Visualizar total de compras por cidade')
print('')
print(df_q11.to_string(index=False))
print('')
print('Q12 - Visualizar total de vendas por período')
print('')
print(df_q12.to_string(index=False))
print('')
print('Q13 - Visualizar valor de vendas por período')
print('')
print(df_q13.to_string(index=False))
print('')
print('Q14 - Visualizar fornecedor com mais vendas')
print('')
print(df_q14.to_string(index=False))
print('')
print('Q15 - Visualizar fornecedor mais rentável')
print('')
print(df_q15.to_string(index=False))
print('')
print('Q16 - Visualizar total de clientes por fornecedor')
print('')
print(df_q16.to_string(index=False))
```
